```python
#Insight 1
plt.figure(figsize=(20,8))
sns.countplot(data=df_data, x="employment", hue="gender")
```

From this plot, we can infer that there is still a gender disparity with respect to the number of men and number of women in many sectors of work. Perhaps measures can be taken to encourage women to take up more roles within these sectors

```python
#insights2

income_by_gender = df_data.groupby(['gender', 'income > 50K']).size().unstack()  #to convert multiindex into dataframe

# Plot the income distribution by gender
income_by_gender.plot(kind='bar', stacked=True)

# Add labels and title to the plot
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Income Distribution by Gender')

# Show the plot
plt.show()
```

The Above resulting visualization shows the income distribution by gender. From the plot, we can see that males have a higher proportion of incomes above 50K than females.

```python
# Insight 3
# Group the data by occupation and education and the number of records in each group and then resetting the index to a default numerical index
edu_by_occupation = df_data.groupby(['job', 'education']).size().reset_index(name='count')

# Create a heatmap to show the distribution of education levels by occupation
sns.heatmap(edu_by_occupation.pivot('job', 'education', 'count'), cmap='Blues')

# Add labels and title to the plot
plt.xlabel('Education')
plt.ylabel('Occupation')
plt.title('Distribution of Education Levels by Occupation')

# Show the plot
plt.show()
```

The Above resulting visualization shows the distribution of education levels by occupation. From the plot, we can see that certain occupations, such as "Prof-specialty" and "Exec-managerial", tend to require higher levels of education. Conversely, occupations such as "Farming-fishing" and "Priv-house-serv" tend to require lower levels of education.
